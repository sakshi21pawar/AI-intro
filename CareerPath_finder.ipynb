```python
def create_professional_ui(self):
    """Create clean, professional question interface"""
    if self.current_question >= len(self.questions):
        self.show_professional_results()
        return

    question = self.questions[self.current_question]

    # Professional header
    header = widgets.HTML(value=f"""
    <div style="background: #2C3E50; padding: 20px; border-radius: 8px; color: white; margin-bottom: 20px;">
        <h3 style="margin: 0; font-weight: 500;">Career Assessment</h3>
        <p style="margin: 5px 0; opacity: 0.9;">Question {self.current_question + 1} of {len(self.questions)}</p>
    </div>
    """)

    # Question
    question_text = widgets.HTML(value=f"""
    <div style="background: #F8F9FA; padding: 20px; border-radius: 8px; border-left: 4px solid #3498DB; margin-bottom: 20px;">
        <h4 style="margin: 0; color: #2C3E50;">{question['text']}</h4>
    </div>
    """)

    # Progress bar
    progress_percent = (self.current_question + 1) / len(self.questions) * 100
    progress_bar = widgets.HTML(value=f"""
    <div style="background: #ECF0F1; height: 6px; border-radius: 3px; margin: 20px 0;">
        <div style="background: #27AE60; width: {progress_percent}%; height: 100%; border-radius: 3px; transition: width 0.3s;"></div>
    </div>
    <div style="text-align: center; color: #7F8C8D; font-size: 14px;">
        {int(progress_percent)}% Complete
    </div>
    """)

    # Option buttons
    option_buttons = []
    for text, trait in question['options']:
        btn = widgets.Button(
            description=text,
            layout=widgets.Layout(width='90%', height='50px', margin='8px 0'),
            style={'button_color': '#FFFFFF', 'font_weight': 'normal'}
        )
        btn.add_class('professional-btn')
        btn.trait = trait
        btn.on_click(self.on_professional_option_click)
        option_buttons.append(btn)

    # Layout
    options_box = widgets.VBox(option_buttons, layout=widgets.Layout(align_items='center'))
    main_layout = widgets.VBox([
        header,
        question_text,
        progress_bar,
        options_box
    ])

    return main_layout

def on_professional_option_click(self, btn):
    """Handle option selection professionally"""
    self.user_traits.append(btn.trait)
    self.current_question += 1

    # Show loading transition
    clear_output()
    display(widgets.HTML(value="""
    <div style="text-align: center; padding: 50px;">
        <div style="color: #3498DB; font-size: 48px; margin-bottom: 20px;">●</div>
        <h3 style="color: #2C3E50;">Processing Response</h3>
        <p style="color: #7F8C8D;">Analyzing your career preferences...</p>
    </div>
    """))

    time.sleep(0.5)

    # Show next question or results
    clear_output()
    if self.current_question < len(self.questions):
        display(self.create_professional_ui())
    else:
        self.show_professional_results()

ProfessionalCareerExpert.create_professional_ui = create_professional_ui
ProfessionalCareerExpert.on_professional_option_click = on_professional_option_click
```

```python
def show_professional_results(self):
    """Display professional results with clean design"""
    # Calculate career matches
    scores = {}
    for career, info in self.careers.items():
        score = sum(1 for trait in info['traits'] if trait in self.user_traits)
        match_percent = (score / len(info['traits'])) * 100
        scores[career] = match_percent

    top_careers = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:3]

    # Results header
    display(HTML("""
    <div style="background: linear-gradient(135deg, #34495E, #2C3E50);
                padding: 30px; border-radius: 8px; color: white; margin-bottom: 30px;">
        <h1 style="margin: 0; font-weight: 400; font-size: 28px;">Career Path Analysis</h1>
        <p style="margin: 10px 0 0 0; opacity: 0.9;">Based on your skills and preferences assessment</p>
    </div>
    """))

    # Top career matches
    display(HTML("""
    <div style="background: #F8F9FA; padding: 20px; border-radius: 8px; margin-bottom: 25px;">
        <h3 style="color: #2C3E50; margin: 0 0 15px 0;">Recommended Career Paths</h3>
    </div>
    """))

    for i, (career, match) in enumerate(top_careers):
        info = self.careers[career]

        # Different border colors for ranking
        border_colors = ['#E74C3C', '#3498DB', '#27AE60']

        display(HTML(f"""
        <div style="border: 2px solid {border_colors[i]}; border-radius: 8px; padding: 0; margin-bottom: 20px; background: white;">
            <div style="background: {border_colors[i]}; color: white; padding: 15px 20px;">
                <h3 style="margin: 0; font-weight: 500;">#{i+1} {career}</h3>
                <div style="font-size: 18px; margin-top: 5px;">Match: {match:.0f}%</div>
            </div>

            <div style="padding: 20px;">
                <div style="margin-bottom: 15px;">
                    <strong style="color: #2C3E50;">Career Description:</strong>
                    <p style="color: #5D6D7E; margin: 8px 0;">{info['description']}</p>
                </div>

                <div style="display: grid; grid-template-columns: repeat(auto-fit, minmax(200px, 1fr)); gap: 15px;">
                    <div>
                        <strong style="color: #2C3E50; display: block;">Salary Range</strong>
                        <span style="color: #27AE60;">{info['salary']}</span>
                    </div>
                    <div>
                        <strong style="color: #2C3E50; display: block;">Job Growth</strong>
                        <span style="color: #2980B9;">{info['growth']}</span>
                    </div>
                    <div>
                        <strong style="color: #2C3E50; display: block;">Education</strong>
                        <span style="color: #8E44AD;">{info['education']}</span>
                    </div>
                </div>

                <div style="margin-top: 15px;">
                    <strong style="color: #2C3E50; display: block;">Key Skills Required</strong>
                    <p style="color: #5D6D7E; margin: 8px 0;">{info['skills']}</p>
                </div>
            </div>
        </div>
        """))

    # Skills Analysis
    self.show_skills_analysis()

    # Next Steps
    self.show_professional_next_steps()

    # Restart option
    restart_btn = widgets.Button(
        description="New Assessment",
        button_style='primary',
        layout=widgets.Layout(width='180px', height='40px')
    )
    restart_btn.on_click(self.restart_assessment)

    display(widgets.HBox([restart_btn], layout=widgets.Layout(justify_content='center', margin='20px 0')))

def show_skills_analysis(self):
    """Display professional skills analysis"""
    trait_counts = {}
    for trait in self.user_traits:
        trait_counts[trait] = trait_counts.get(trait, 0) + 1

    # Map traits to skill categories
    skill_categories = {
        'Technical Skills': ['tech_lover', 'problem_solver', 'logical_thinker'],
        'Analytical Skills': ['analytical_mind', 'numbers_lover', 'science_lover'],
        'Creative Skills': ['creative', 'artistic', 'visual_thinker'],
        'Interpersonal Skills': ['social_butterfly', 'helper', 'communicator', 'leader']
    }

    display(HTML("""
    <div style="background: #F8F9FA; padding: 20px; border-radius: 8px; margin: 25px 0;">
        <h3 style="color: #2C3E50; margin: 0 0 15px 0;">Skills Profile Analysis</h3>
    </div>
    """))

    skills_html = """
    <div style="background: white; border-radius: 8px; padding: 20px; border: 1px solid #EAECEE;">
    """

    for category, traits in skill_categories.items():
        score = sum(1 for trait in traits if trait in self.user_traits)
        max_score = len(traits)
        percentage = (score / max_score) * 100

        # Determine proficiency level
        if percentage >= 75:
            level = "Advanced"
            color = "#27AE60"
        elif percentage >= 50:
            level = "Proficient"
            color = "#3498DB"
        elif percentage >= 25:
            level = "Developing"
            color = "#F39C12"
        else:
            level = "Beginner"
            color = "#E74C3C"

        skills_html += f"""
        <div style="margin-bottom: 20px;">
            <div style="display: flex; justify-content: between; margin-bottom: 8px;">
                <strong style="color: #2C3E50;">{category}</strong>
                <span style="color: {color}; font-weight: 500;">{level}</span>
            </div>
            <div style="background: #ECF0F1; height: 8px; border-radius: 4px;">
                <div style="background: {color}; width: {percentage}%; height: 100%; border-radius: 4px;"></div>
            </div>
            <div style="display: flex; justify-content: between; margin-top: 5px;">
                <span style="color: #7F8C8D; font-size: 12px;">{score}/{max_score} competencies</span>
                <span style="color: #7F8C8D; font-size: 12px;">{percentage:.0f}%</span>
            </div>
        </div>
        """

    skills_html += "</div>"
    display(HTML(skills_html))

def show_professional_next_steps(self):
    """Display professional next steps"""
    display(HTML("""
    <div style="background: #F8F9FA; padding: 20px; border-radius: 8px; margin: 25px 0;">
        <h3 style="color: #2C3E50; margin: 0 0 15px 0;">Recommended Action Plan</h3>
    </div>
    """))

    steps = [
        ("Research Phase", "Conduct detailed research on top career matches, including job requirements and market trends"),
        ("Skill Development", "Identify and begin developing key skills through online courses and practical projects"),
        ("Networking", "Connect with professionals in target industries through LinkedIn and professional events"),
        ("Education Planning", "Research educational requirements and identify relevant degree programs or certifications"),
        ("Portfolio Building", "Develop projects and build a portfolio that demonstrates your skills and capabilities"),
        ("Career Preparation", "Prepare application materials and practice for interviews in your target field")
    ]

    steps_html = """
    <div style="background: white; border-radius: 8px; padding: 20px; border: 1px solid #EAECEE;">
    """

    for i, (phase, action) in enumerate(steps, 1):
        steps_html += f"""
        <div style="display: flex; margin-bottom: 15px; padding-bottom: 15px; border-bottom: 1px solid #F8F9FA;">
            <div style="background: #3498DB; color: white; width: 24px; height: 24px; border-radius: 50%;
                     display: flex; align-items: center; justify-content: center; font-size: 12px; margin-right: 15px;">
                {i}
            </div>
            <div>
                <strong style="color: #2C3E50; display: block;">{phase}</strong>
                <span style="color: #5D6D7E;">{action}</span>
            </div>
        </div>
        """

    steps_html += "</div>"
    display(HTML(steps_html))

def restart_assessment(self, btn):
    """Restart the assessment"""
    self.user_traits = []
    self.current_question = 0
    clear_output()
    self.start_professional_assessment()

ProfessionalCareerExpert.show_professional_results = show_professional_results
ProfessionalCareerExpert.show_skills_analysis = show_skills_analysis
ProfessionalCareerExpert.show_professional_next_steps = show_professional_next_steps
ProfessionalCareerExpert.restart_assessment = restart_assessment
```
